```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Constants
M = 2.23
Ln = 5.32
d = 5
df = 5
dr = 4
lt = 1.41
xp = 9.21
cr = np.arange(2,5,0.01)
ct = 0
S = np.arange(2,5,0.01)
R = 2
xr = cr
xb = 9.21
n = 4 # number of fins

```

```python
def lf(cr,S):
    return (S**2+(cr**2)/4)**(0.5)
```

```python
cnn = 2/np.sqrt(M**2-1)
xn = 0.466*Ln
cnt =  2*((dr/d)**2-(df/d)**2)
xt = xp + (lt/3)*(1+ (1-df/dr)/(1-(df/dr)**2))
# cnf = (1+R/(S+R))*(4*n*(S/d)**2/(1+np.sqrt(1+(2*lf(cr,S)/(cr+ct))**2 )))
# xf = xb + (xr/3)* (cr+2*ct)/(cr+ct) + (1/6)*((cr+ct)-cr*ct/(cr+ct))
cnf = np.zeros((len(cr),len(cr)))
xf = np.zeros((len(cr),len(cr)))
for i in range(0,len(cr)):
    for j in range(0,len(cr)):
        cnf[i][j] = (1+R/(S[i]+R))*(4*n*(S[i]/d)**2/(1+np.sqrt(1+(2*lf(cr[j],S[i])/(cr[j]+ct))**2 )))
        xf[i][j] = xb + (xr[j]/3)* (cr[j]+2*ct)/(cr[j]+ct) + (1/6)*((cr[j]+ct)-cr[j]*ct/(cr[j]+ct))
```

```python
cnr = cnn +cnt +cnf
xcp = (cnn*xn + cnt*xt + cnf*xf)/cnr
```

```python
plt.plot(cr,xcp[:,])
plt.grid()
```

```python
plt.plot(S, xcp)
plt.grid()
```

```python
plt.contour(cr, S, xcp, 100)
plt.colorbar()
plt.grid()
```
